```python
import cv2
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
import cv2

img = cv2.imread('cc0690c1fc34726a2b25.jpg')

# Lưu hình ảnh với định dạng khác (.png)
cv2.imwrite('image_converted.png', img)
    
# hiển thị ảnh 
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
plt.imshow(img_rgb)
plt.axis('off')  
plt.show()
```

```python
# Chuyển sang Grayscale (Ảnh xám)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Chuyển sang HSV (Hue, Saturation, Value)
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Chuyển sang LAB
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

#hiển thị ảnh 3 kênh màu 
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(gray, cmap='gray')
axs[0].set_title('Grayscale')
axs[0].axis('off')
axs[1].imshow(hsv)
axs[1].set_title('HSV')
axs[1].axis('off')
axs[2].imshow(lab)
axs[2].set_title('LAB')
axs[2].axis('off')
plt.show()

```

```python
# Cắt một vùng ảnh (y từ 50 đến 200, x từ 100 đến 300)
roi = img[50:200, 100:300]

# Thay đổi kích thước cố định 
resized_fixed = cv2.resize(img, (400, 400))

# Thay đổi theo tỷ lệ
resized_ratio = cv2.resize(img, None, fx=0.5, fy=0.5)

# hiển thị ảnh đã cắt và thay đổi kích thước 
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB))
axs[0].set_title('Cropped Image')
axs[0].axis('off')
axs[1].imshow(cv2.cvtColor(resized_fixed, cv2.COLOR_BGR2RGB))
axs[1].set_title('Resized (Fixed)')
axs[1].axis('off')
axs[2].imshow(cv2.cvtColor(resized_ratio, cv2.COLOR_BGR2RGB))
axs[2].set_title('Resized (Ratio)')
axs[2].axis('off')
plt.show()
```

```python
# Tạo một bản sao để vẽ
canvas = img.copy()

# Vẽ đường thẳng: (ảnh, điểm đầu, điểm cuối, màu BGR, độ dày)
cv2.line(canvas, (0, 0), (200, 200), (255, 0, 0), 3)

# Vẽ hình tròn: (ảnh, tâm, bán kính, màu, độ dày; -1 là tô đặc)
cv2.circle(canvas, (150, 150), 50, (0, 255, 0), 2)

# Vẽ hình chữ nhật: (ảnh, góc trên trái, góc dưới phải, màu, độ dày)
cv2.rectangle(canvas, (50, 50), (250, 250), (0, 0, 255), 2)

# Thêm văn bản
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(canvas, 'Gemini AI - OpenCV', (10, 30), font, 1, (255, 255, 255), 2)

# hiển thị ảnh đã vẽ
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
